```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, OneHotEncoder
```

```python
df = pd.read_csv('forestfires.csv')
```

```python
X = df.drop('target', axis=1)
y = df['target']
```

```python
cat_col = ['month', 'day']
```

```python
enc = OneHotEncoder(sparse_output=False)
cat_encoded = enc.fit_transform(X[cat_col])
```

```python
enc_df = pd.DataFrame(cat_encoded, columns=enc.get_feature_names_out(cat_col))
```

```python
X = X.drop(cat_col, axis=1)
X = pd.concat([X, enc_df], axis=1)
```

```python
ss = StandardScaler()
X_scaled = ss.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
model = LogisticRegression(max_iter=500)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
acc = accuracy_score(y_test, y_pred)
print(f'Accuracy: {acc}')
```

```python
coftrix = confusion_matrix(y_test, y_pred)
print('Confusion Matrix:', coftrix)
```

```python
plt.figure(figsize=(8, 6))
plt.imshow(coftrix, cmap=plt.cm.Oranges)
plt.title('Confusion Matrix')
plt.colorbar()
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.xticks([0, 1], ['Non-occurrence', 'Occurrence'])
plt.yticks([0, 1], ['Non-occurrence', 'Occurrence'])
plt.show()
```
